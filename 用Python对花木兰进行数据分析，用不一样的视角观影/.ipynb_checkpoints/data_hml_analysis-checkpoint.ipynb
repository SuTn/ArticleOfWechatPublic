```python
import pandas as pd
import numpy as np
import pyecharts
from pyecharts.charts import wordcloud
from pyecharts import Pie, Bar
# 只需要在顶部声明 CurrentConfig.ONLINE_HOST 即可
# from pyecharts.globals import CurrentConfig
# CurrentConfig.ONLINE_HOST = "http://127.0.0.1:8000/assets/"
from collections import Counter
```

```python
df = pd.read_csv('./data/data_hml.csv')
print(df.describe())
```

```python
print(df.columns)
```

```python
score_list = list(np.linspace(0, 5, 11))
scores = {}
for i in score_list:
    scores["{}".format(i)] = list(df.score).count(i)
print(sum(list(scores.values())))
scores = sorted(scores.items(),key=lambda x:x[0])
```

```python
scores
```

```python
bar = Bar("花木兰")
bar.add("评分分布", list(i[0] for i in scores), 
        list(i[1] for i in scores),is_more_utils=True,
        is_label_show=True, is_convert=True, label_pos='right')
bar.render("score.html")
```

```python
# -*- coding:utf-8 -*-  
#下面是评论者的城市分布
def remove_not_exist_city(city_list):

    with open(
            r'D:\ProgramData\Anaconda3\envs\dl\Lib\site-packages\pyecharts\datasets\city_coordinates.json',
            mode='r', encoding='utf-8') as f:
        text = f.read()
        if text.startswith(u'\ufeff'):
            text = text.encode('utf8')[3:].decode('utf8')
#         data = json.loads(f.read())  # 将str转换为json
        data = json.loads(text)  # 将str转换为json



    data_new = data.copy()  
    for city in list(i[0] for i in city_list):  
        count = 0
        for i in data.keys():
            count += 1
            if i == city:
                break
            elif i.startswith(city):  # 处理一些地址简写的地名
#                 print(i, city)
                data_new[city] = data[i]
                break
            elif i.startswith(city[0:-1]) and len(city) >= 3:  # 有的地方县改为市区的情况，进行处理
#                 print(i, city)
                data_new[city] = data[i]
                break
        # 如果地名还是不存在的话，就将该地址删除
        if count == len(data):
            for i in city_list:
                if i[0] == city:
                    city_list.remove(i)
    with open(
            r'D:\ProgramData\Anaconda3\envs\dl\Lib\site-packages\pyecharts\datasets\city_coordinates.json',
            mode='w', encoding='utf-8') as f:
        f.write(json.dumps(data_new, ensure_ascii=False))  # 将json转换为str
        
    return city_list


from pyecharts import Geo
import json
cityname = df['cityName']
cityname = cityname.dropna()
city_dict = Counter(cityname)
city_dict = dict(city_dict)
city_list = []
for i in city_dict.items():
    city_list.append((i[0], i[1]))

print(len(city_list))
```

```python
city_list = remove_not_exist_city(city_list)
print(len(city_list))
city_list = sorted(city_list, key=lambda x:x[1], reverse=True)
print(city_list[:100])
```
